```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import silhouette_score
import time
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
X, y = make_moons(n_samples = 5000, noise = 0.1)
np.shape(X), np.shape(y)
```

```python
np.unique(y)
```

```python
plt.rcParams["figure.figsize"] = [10, 10]
data_df = pd.DataFrame(X, columns = ['xaxis', 'yaxis'])
sns.scatterplot(data=data_df, x="xaxis", y="yaxis", style= y #, style="time"
               )
```

```python
## Applying DBSCAN to this
dbscan = DBSCAN(eps = 0.07, min_samples = 5)
dbscan.fit(X)
```

```python
dbscan.labels_  ## cluster labels are stored in this variable
```

```python
np.unique(dbscan.labels_)
```

```python
dbscan.core_sample_indices_, len(dbscan.core_sample_indices_,) ## indices of the core points
```

```python
dbscan.components_ , len(dbscan.components_)
```

```python
sns.scatterplot(data=data_df, x="xaxis", y="yaxis", hue= dbscan.labels_ , style=y)
```

```python
# How do we choose the right eps and min_pts?
```

```python
## Choosing min_pts = 4 since no of dimensions = 2

# for min pts
from sklearn.neighbors import NearestNeighbors
from matplotlib import pyplot as plt

neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)
```

```python
distances.shape
```

```python
indices.shape
```

```python
distances[:4]
```

```python
indices[:4]
```

```python
distances_1 = np.mean(distances[:,1:],axis=1)
distances_2 = np.sort(distances_1, axis=0)
plt.plot(distances_2)
```

```python
## here we are plotting avg distance of each point to it's k nearest neighbours
# choose the inflexion pt as optimal epsilon
```

```python
dbscan = DBSCAN(eps = 0.04, min_samples = 5)
dbscan.fit(X)
```

```python
np.unique(dbscan.labels_)
```

```python
sns.scatterplot(data=data_df, x="xaxis", y="yaxis", hue= dbscan.labels_ , style=y)
```

# Predictions

```python
X_new = np.array([[-0.5,0], [0,0.5], [1,-0.1],[2,1]])
data_new_df = pd.DataFrame(X_new, columns = ['xaxis','yaxis'])
sns.scatterplot(data=data_df, x="xaxis", y="yaxis", hue= dbscan.labels_ , style=y)
sns.scatterplot(data=data_new_df, x="xaxis", y="yaxis")
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 20)
knn.fit(dbscan.components_, dbscan.labels_[dbscan.core_sample_indices_]) ## using only core instances
```

```python
y_dist, y_pred_idx = knn.kneighbors(X_new, n_neighbors = 1)
np.shape(y_dist), np.shape(y_pred_idx) 
```

```python
y_dist
```

```python
y_pred_idx
```

```python
y_pred = dbscan.labels_[dbscan.core_sample_indices_][y_pred_idx] 
## Since we trained only on core points using core indices
```

```python
y_pred
```

```python
y_pred[y_dist > 0.05] = -1
y_pred
```

# Hyperparamters

### class sklearn.cluster.DBSCAN(eps=0.5, *, min_samples=5, metric='euclidean', metric_params=None, algorithm='auto', leaf_size=30, p=None, n_jobs=None)

### metric can be any of pairwise distances 

```python
# epsilon and min_samples are the major hyper params

# use min_samples based on no. of dimensions
# Use elbow curve for choosing epsilon
```

```python
## Different values of min_samples

min_samples = [3, 5, 7, 10, 15, 20, 30, 40]
fig, ax = plt.subplots(4, 2, figsize=(20,20))
for i in range(len(min_samples)):
    dbscan = DBSCAN(eps = 0.05, min_samples = min_samples[i])
    dbscan.fit(X)
    q, mod = divmod(i, 2)
    sns.scatterplot(data = data_df, x="xaxis", y="yaxis", hue= dbscan.labels_ ,\
        style = y,ax=ax[q][mod]).set(title =  "Min Samples = " +\
                                     str(min_samples[i]) + " , " + str(len(np.unique(dbscan.labels_))) + " Clusters")
```

```python
## Different values of min_samples

eps = [0.01, 0.05, 0.1, 0.2, 0.5, 1]
fig, ax = plt.subplots(3, 2, figsize=(20,20))
for i in range(len(eps)):
    dbscan = DBSCAN(eps = eps[i], min_samples = 5)
    dbscan.fit(X)
    q, mod = divmod(i, 2)
    sns.scatterplot(data = data_df, x="xaxis", y="yaxis", hue= dbscan.labels_ ,\
        style = y,ax=ax[q][mod]).set(title =  "Epsilon = " +\
                                     str(eps[i]) + " , " + str(len(np.unique(dbscan.labels_))) + " Clusters")
```
