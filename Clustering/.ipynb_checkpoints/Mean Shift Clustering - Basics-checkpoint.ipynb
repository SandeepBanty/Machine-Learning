# Application in Python

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import silhouette_score
import time
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
## Making 5 2D blobs
X, y, centers = make_blobs(n_samples=2000, centers=5, n_features=2,random_state=0,cluster_std = 0.5, return_centers = True)
np.shape(X), np.shape(y), np.shape(centers)
```

```python
plt.rcParams["figure.figsize"] = [10, 10]
data_df = pd.DataFrame(X, columns = ['xaxis', 'yaxis'])
sns.scatterplot(data=data_df, x="xaxis", y="yaxis", style= y 
               )
```

class sklearn.cluster.MeanShift(*, bandwidth=None, seeds=None, bin_seeding=False, min_bin_freq=1, cluster_all=True, n_jobs=None, max_iter=300)

```python
from sklearn.cluster import MeanShift
```

```python
ms = MeanShift(bandwidth = 1, cluster_all = False)
ms.fit(X)
```

```python
sns.scatterplot(data=data_df, x="xaxis", y="yaxis", hue= ms.labels_)
```

```python
ms.labels_
```

```python
## How do we choose the right bandwidth (R)?
```

```python
from sklearn.cluster import  estimate_bandwidth

bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=500)
bandwidth

### Consider pairwise distances between all 500 samples and taking 20th percentile of that as bandwidth
## why not 30th percentile?
```

```python
ms = MeanShift(bandwidth = bandwidth, cluster_all = False)
ms.fit(X)
```

```python
sns.scatterplot(data=data_df, x="xaxis", y="yaxis", hue= ms.labels_)
```

```python
### Does this work on non spherical clusters
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=2000, n_features=2, n_redundant=0, random_state=4)
```

```python
# Plotting the blobs using a scatter plot
plt.rcParams["figure.figsize"] = [15, 15]
plt.scatter(X[:,0],X[:,1])
```

```python
bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=500)
bandwidth
```

```python
ms = MeanShift(bandwidth = bandwidth, cluster_all = False)
ms.fit(X)
```

```python
data = pd.DataFrame(X, columns = ['x1','x2'])
data['pred'] = ms.labels_
sns.scatterplot(data=data, x="x1", y="x2", hue="pred")
```
