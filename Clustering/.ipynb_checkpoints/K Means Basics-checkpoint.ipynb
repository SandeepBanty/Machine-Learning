```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans,MiniBatchKMeans
from scipy.spatial import Voronoi, voronoi_plot_2d
import pandas as pd
from sklearn.metrics import silhouette_score
import time
from yellowbrick.cluster import KElbowVisualizer
import seaborn as sns
```

```python
## Making 5 2D blobs
X, y, centers = make_blobs(n_samples=2000, centers=5, n_features=2,random_state=0,cluster_std = 0.5, return_centers = True)
np.shape(X), np.shape(y), np.shape(centers)
```

```python
# Plotting the blobs using a scatter plot
plt.rcParams["figure.figsize"] = [15, 15]
plt.scatter(X[:,0],X[:,1])
plt.scatter(centers[:,0],centers[:,1], marker='s', s=300, color = 'r')
```

## sklearn kmeans documentation

### 1. sklearn.cluster.KMeans(n_clusters=8, *, init='k-means++', n_init='warn', max_iter=300, tol = 0.0001, verbose=0, random_state=None, copy_x=True, algorithm='lloyd')
### 2. default n_init is 10
### 3. tol is relative tolerance with regards to Frobenius norm of the difference in the cluster centers of two consecutive iterations to declare convergence.
### 4. algorithm - llyod is traditional kmeans and elkan uses triangular inequality to reduce no of distance calculations (sometimes useful when no. of clusters is very high)

### Applying k means to this dataset

```python
kmeans = KMeans(n_clusters = 5, random_state = 5)
y_pred = kmeans.fit(X)
```

```python
kmeans.labels_ , kmeans.cluster_centers_, kmeans.inertia_, silhouette_score(X,kmeans.labels_)
```

```python
# kmeans.transform gives distance between every point and every cluster
kmeans.transform(X).round(2)
```

```python
# Plotting decision boundaries to understand better

vor = Voronoi(kmeans.cluster_centers_)
fig = voronoi_plot_2d(vor,plt.gca())
plt.scatter(X[:,0],X[:,1], color = 'y')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], marker='s', s=300, color = 'r')
plt.xlim([-3, 11])
plt.ylim([-5, 9])

plt.show()
```

```python
# What if we set n_init to 1 and randomize the initialization

kmeans_2 = KMeans(n_clusters = 5, n_init = 1, init = 'random')
y_pred = kmeans_2.fit(X)


# Plotting decision boundaries to understand better
vor = Voronoi(kmeans_2.cluster_centers_)
fig = voronoi_plot_2d(vor,plt.gca())
plt.scatter(X[:,0],X[:,1], color = 'y')
plt.scatter(kmeans_2.cluster_centers_[:,0],kmeans_2.cluster_centers_[:,1], marker='s', s=300, color = 'r')
plt.xlim([-3, 11])
plt.ylim([-5, 9])

plt.show()

## We get really bad clusters for some initializations
```

```python
kmeans_2.inertia_ , silhouette_score(X,kmeans_2.labels_)
```

### A few variants of KMeans - comparing time taken, inertia and silhouette score

```python
# Comparing kmeans, mini-batch kmeans and accelerated kmeans

kmeans_1 = KMeans(n_clusters = 5, random_state = 5)
minibatch = MiniBatchKMeans(n_clusters = 5, random_state = 5, batch_size = 1024)
kmeans_accelerated = KMeans(n_clusters = 5, random_state = 5, algorithm = 'elkan')

performance_summary = pd.DataFrame(columns = ['model', 'inertia', 'silhouette_score', 'time_taken'])
models = {'kmeans_1' : kmeans_1,'minibatch': minibatch,'kmeans_accelerated': kmeans_accelerated}
```

```python
for key in models.keys():
# store starting time
    model = models[key]
    begin = time.time()
    y = model.fit(X)
    end = time.time()
    inertia = round(model.inertia_,3)
    ssc = round(silhouette_score(X,model.labels_),3)
    time_taken = round(end - begin,3)
    performance_summary.loc[len(performance_summary)] = [key, inertia , ssc , time_taken]

performance_summary
```

### Finding the optimal value of K using elbow curve and silhouette score

```python
# Instantiate the clustering model and visualizer
km = KMeans(random_state=42)
visualizer = KElbowVisualizer(km, k=(2,15))

visualizer.fit(X)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
  
fig, ax = plt.subplots(4, 2, figsize=(20,20))
for i in [2, 3, 4, 5, 6, 7 , 8, 9]:
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=42)
    q, mod = divmod(i, 2)
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(X) 
```
