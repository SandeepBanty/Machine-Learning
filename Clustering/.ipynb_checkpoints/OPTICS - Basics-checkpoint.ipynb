# Application in Python

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import silhouette_score
import time
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
## Making 5 2D blobs
X, y, centers = make_blobs(n_samples=2000, centers=5, n_features=2,random_state=0,cluster_std = 0.5, return_centers = True)
np.shape(X), np.shape(y), np.shape(centers)
```

```python
plt.rcParams["figure.figsize"] = [10, 10]
data_df = pd.DataFrame(X, columns = ['xaxis', 'yaxis'])
sns.scatterplot(data=data_df, x="xaxis", y="yaxis", style= y)
```

## sklearn hyperparams

class sklearn.cluster.OPTICS(*, min_samples=5, max_eps=inf, metric='minkowski', p=2, metric_params=None, cluster_method='xi', eps=None, xi=0.05, predecessor_correction=True, min_cluster_size=None, algorithm='auto', leaf_size=30, memory=None, n_jobs=None)

```python
from sklearn.cluster import OPTICS
```

```python
op = OPTICS(min_samples=50, xi=0.1, min_cluster_size=0.05)
op.fit(X)
```

```python
sns.scatterplot(data=data_df, x="xaxis", y="yaxis", hue= op.labels_)
```

```python
## Plotting the reachability
reachability = op.reachability_[op.ordering_]
labels = op.labels_[op.ordering_]

reachability

op.ordering_

np.arange(len(X))

space = np.arange(len(X))
colors = ["g.", "r.", "b.", "y.", "c."]
plt.figure(figsize=(20, 15))

for klass, color in zip(range(0, 5), colors):
    Xk = space[labels == klass]
    Rk = reachability[labels == klass]
    plt.plot(Xk, Rk, color, alpha=0.3)
```

```python
from sklearn import cluster, datasets, mixture
X,y = datasets.make_blobs(
    n_samples=2000, cluster_std=[1.0, 2.5, 0.5], random_state=25
)
```

```python
# Plotting the blobs using a scatter plot
plt.rcParams["figure.figsize"] = [15, 15]
plt.scatter(X[:,0],X[:,1])
```

```python
## trying DBSCAN on this dataset

from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 0.1, min_samples = 5)
dbscan.fit(X)
```

```python
dbscan.labels_  ## cluster labels are stored in this variable
```

```python
sns.scatterplot(data=None, x = X[:,0], y = X[:,1], hue= dbscan.labels_)
```

```python

eps = [0.01, 0.05, 0.1, 0.2, 0.5, 1]
fig, ax = plt.subplots(3, 2, figsize=(20,20))
for i in range(len(eps)):
    dbscan = DBSCAN(eps = eps[i], min_samples = 5)
    dbscan.fit(X)
    q, mod = divmod(i, 2)
    sns.scatterplot(data = None, x = X[:,0], y = X[:,1], hue= dbscan.labels_ ,ax=ax[q][mod]).set(title =  "Epsilon = " +\
                                     str(eps[i]) + " , " + str(len(np.unique(dbscan.labels_))) + " Clusters")
```

```python
### DBSCAN is not choosing the right clusters for any value of epsilon
```

```python
op = OPTICS(min_samples = 5, xi=0.05)
op.fit(X)
```

```python
sns.scatterplot(data=None, x = X[:,0], y = X[:,1], hue= op.labels_)
```

```python
## choosing cluster using xi is giving bad clusters. 
## Let's look at the reachability plot and manually decide thresholds for each cluster
```

```python
## Plotting the reachability
reachability = op.reachability_[op.ordering_]
labels = op.labels_[op.ordering_]
y_ordered = y[op.ordering_]
```

```python
reachability
op.ordering_

space = np.arange(len(X))
space
```

```python
min(reachability), max(reachability)
```

```python
colors = ["g.", "r.", "b.", "y."]
plt.figure(figsize=(20, 10))
plt.plot(space, reachability, alpha=0.8)

plt.xticks(np.arange(min(space), max(space)+1, 50))
```
