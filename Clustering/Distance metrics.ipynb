## This notebook covers different distance metrics and the intuitions behind them

Lets say there are 2 two points $\overrightarrow A$ and $\overrightarrow B$ in an ***n*** dimensional space. 

$$ Euclidean(\overrightarrow A,\overrightarrow B)  \quad =   \quad \sqrt{\sum_{i=0}^{n}(A_i^2 - B_i^2)} $$
 
$$ Manhattan(\overrightarrow A,\overrightarrow B)  \quad =   \quad \sum_{i=0}^{n}|A_i - B_i| $$

$$ Cosine Distance (\overrightarrow A,\overrightarrow B) \quad = \quad 1 - Cosine Similarity (\overrightarrow A,\overrightarrow B) \newline
                             \quad  \quad \quad \quad \: \: \: =  \quad 1 - \dfrac{\overrightarrow A - \overrightarrow B}{||\overrightarrow A|||\overrightarrow B||}  $$



     

## What do these distances even mean?

Let's say we join two points in 2D space using a right angled triangle. 

1. Manhattan distance is the sum of the two sides of the triangle. Same as L1 distance or city block distance
2. Eucliean distance is the length of the hypotenuse of the triangle. Same as L2 distance
3. Cosine Distance is 1 - cosine of angle between the vectors of two points. This measures only the angular separation between the two vectors and doesn't care about the direct distance
