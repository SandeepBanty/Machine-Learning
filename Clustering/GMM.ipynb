```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
import seaborn as sns
from sklearn.mixture import GaussianMixture
```

```python
X, y = make_classification(n_samples=2000, n_features=2, n_redundant=0, random_state=4)
```

```python
# Plotting the blobs using a scatter plot
plt.rcParams["figure.figsize"] = [15, 15]
plt.scatter(X[:,0],X[:,1])
```

```python
gm = GaussianMixture(n_components = 4, n_init = 10)
gm.fit(X)
```

```python
# GMM is a generative model. It can produce new data
X_new, y_new = gm.sample(10)
X_new
```

```python
# To estimate the density of the model at any given location

gm.score_samples(X).round(2)
```

```python
data = pd.DataFrame(X, columns = ['x1','x2'])
data['pred'] = gm.predict(X)
sns.scatterplot(data=data, x="x1", y="x2", hue="pred")
```

```python
# choosing the constraint on covariance matrix
types = ['spherical', 'diag', 'tied', 'full']

fig, ax = plt.subplots(2, 2, figsize=(20,20))
for i in range(len(types)):
    gm = GaussianMixture(n_components = 4, n_init = 10, covariance_type = types[i])
    gm.fit(X)
    labels = gm.predict(X)
    q, mod = divmod(i, 2)
    sns.scatterplot(data = data, x="x1", y="x2", hue= labels, ax=ax[q][mod]).set(title =  "Covariance Type = " + str(types[i]))
    
```

```python
## Only full was able to detect the right clusters. But it's slower to run
```

```python
# Choosing the right number of clusters
```

```python
# choosing the constraint on covariance matrix
no_of_clusters = [2,3,4,5,6,8,10,15]
bic = []
for i in range(len(no_of_clusters)):
    gm = GaussianMixture(n_components = no_of_clusters[i], n_init = 10, covariance_type = 'full')
    gm.fit(X)
    labels = gm.predict(X)
    bic.append(gm.bic(X))
    
```

```python
## Plotting BIC vs no of clusters
sns.lineplot(data=None, x=no_of_clusters, y=bic, markers=True)
```

```python
# 4 is the right no. of clusters
```

# Drawbacks of GMM

1. Real world data is not always gaussian
2. Can't work with categorical variables
3. Need to specify the no. of clusters
4. Need sufficient data for each cluster do define a gaussian
