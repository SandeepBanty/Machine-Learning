**BIRCH - Balanced Iterative Reducing and Clustering using Hierarchies**

BIRCH uses clustering feature tree and clustering feature to perform hierarchical clustering. 

1. What is a clustering feature? <br />
  A clustering feature is a triple(3 measures) summarizing the information about a cluster. Given N points(with d dimensions) in a cluster, a clustering feature is written as (N, $\bar {LS}$, SS) <br />
  N - number of data points in the cluster <br />
  LS - Linear sum of the N points i.e. <br />
  SS - Squared sum of the N points <br />
  
  <br />
  
2. What is a clustering feature tree? <br />
  CF tree is a height balanced tree with two parameters branching factor B and threshold T. In regular trees, a node contains a data point whereas in CF trees a node is box of B clusters (B CFs)

# Application in Python

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import silhouette_score
import time
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
## Making 5 2D blobs
X, y, centers = make_blobs(n_samples=2000, centers=5, n_features=2,random_state=0,cluster_std = 0.5, return_centers = True)
np.shape(X), np.shape(y), np.shape(centers)
```

```python
plt.rcParams["figure.figsize"] = [10, 10]
data_df = pd.DataFrame(X, columns = ['xaxis', 'yaxis'])
sns.scatterplot(data=data_df, x="xaxis", y="yaxis", style= y 
               )
```
